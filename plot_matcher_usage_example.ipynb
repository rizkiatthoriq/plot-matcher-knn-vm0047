```python
from plot_matcher import PlotMatcherKNN
import pandas as pd
```

```python
# Import the datasets
pp = pd.read_csv('data/Project Plot Points.csv')
cp = pd.read_csv('data/Control Plot Points.csv')
```

```python
# Check the datasets
display(pp.head(), cp.head())
```

The tables contain 2 column category: plot id (PP ID or CP ID) and SI years (2015 to 204)

```python
# Define year columns as strings
project_start_year = 2025 
historical_si_years = [str(year) for year in range(project_start_year - 10, project_start_year)]
```

```python
historical_si_years
```

```python
# 1. Matcher instalation
matcher = PlotMatcherKNN(
    k=5, 
    metric='mahalanobis', 
    use_scaler=False, # Set to False based on your data characteristics
    id_col_labeled='PP ID',
    id_col_unlabeled='CP ID'
)
```

```python
# 2. fit_transform to train model and finds pair, it will use df_pp as labelled data to find matches
matches_df = matcher.fit_transform(
    df_labeled= pp, 
    df_unlabeled= cp, 
    historical_si_cols=historical_si_years
)

print("\n--- Final Matches Found ---")
matches_df.head()
```
